```python
# Use the Whisper OpenAI model to convert the Audio to Text
# If you'd prefer to use an Open Source model, class student Youssef has contributed an open source version
# which I've added to the bottom of this colab

audio_file = open(audio_filename, "rb")  # Otwiera plik audio w trybie binarnym (odczyt). 'audio_filename' to zmienna zawierająca ścieżkę do pliku audio.
transcription = openai.audio.transcriptions.create(  # Używa modelu Whisper z OpenAI do transkrypcji audio na tekst.
    model=AUDIO_MODEL,  # 'AUDIO_MODEL' to zmienna zawierająca nazwę modelu OpenAI, który będzie użyty do transkrypcji (np. Whisper).
    file=audio_file,  # Przesyła otwarty plik audio do API OpenAI.
    response_format="text"  # Określa, że odpowiedź powinna być zwrócona w formie tekstowej (nie jako np. JSON).
)
print(transcription)  # Wypisuje wynik transkrypcji na ekranie. 'transcription' zawiera tekst uzyskany z audio.
```

```python
system_message = "You are an assistant that produces minutes of meetings from transcripts, with summary, key discussion points, takeaways and action items with owners, in markdown."
user_prompt = f"Below is an extract transcript of a Denver council meeting. Please write minutes in markdown, including a summary with attendees, location and date; discussion points; takeaways; and action items with owners.\n{transcription}"

messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
  ]
```

```python
# Konfiguracja kwantyzacji dla modelu w celu zmniejszenia zużycia pamięci i poprawy wydajności obliczeniowej

quant_config = BitsAndBytesConfig(
    load_in_4bit=True,  # Wczytanie modelu w formacie 4-bitowym, co zmniejsza zapotrzebowanie na pamięć
    bnb_4bit_use_double_quant=True,  # Użycie podwójnej kwantyzacji dla jeszcze większej redukcji pamięci
    bnb_4bit_compute_dtype=torch.bfloat16,  # Typ danych używany do obliczeń (bfloat16 pozwala na dobrą precyzję przy mniejszym zużyciu pamięci)
    bnb_4bit_quant_type="nf4"  # Typ kwantyzacji NF4 (Normalized Float 4), który zapewnia wysoką jakość w niskiej precyzji
)
```

```python
# Wczytanie tokenizera dla wybranego modelu LLM (Large Language Model)
tokenizer = AutoTokenizer.from_pretrained(LLAMA)

# Ustawienie tokena paddingu na token końca sekwencji (EOS)
# - jest to ważne dla zgodności modelu w przypadku różnej długości wejść
tokenizer.pad_token = tokenizer.eos_token

# Przetworzenie wiadomości wejściowych na tensor PyTorch (return_tensors="pt")
# i przeniesienie ich na GPU ("cuda") w celu przyspieszenia inferencji
inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")

# Inicjalizacja obiektu do strumieniowego generowania tekstu - pozwala na płynne wyświetlanie wyników
streamer = TextStreamer(tokenizer)

# Wczytanie modelu językowego w trybie mapowania urządzenia "auto" (automatyczne dopasowanie do CPU/GPU)
# oraz z zastosowaniem wcześniej skonfigurowanej kwantyzacji w celu zmniejszenia zużycia pamięci
model = AutoModelForCausalLM.from_pretrained(LLAMA, device_map="auto", quantization_config=quant_config)

# Wygenerowanie odpowiedzi przez model na podstawie wejściowych danych tekstowych
# - `max_new_tokens=2000` oznacza maksymalną liczbę nowych tokenów do wygenerowania
# - `streamer=streamer` umożliwia wyświetlanie tekstu w czasie rzeczywistym zamiast czekać na pełną odpowiedź
outputs = model.generate(inputs, max_new_tokens=2000, streamer=streamer)
```

```python
response = tokenizer.decode(outputs[0])
```
